## Employment 

```python
data = pd.read_csv("Neighborhoods_with_SNAP_Data.csv", index_col = "Neighborhood_2010_HOOD",parse_dates=True) 
#set neighboorhood column as the index
data.index.name = 'Neighborhoods'

jobdata = data[['Total___Jobs_Located_in_N_hood_']] 
jobdata.columns = {'Jobs'}
jobdata.index.name = 'Neighborhoods'
jobdata.head()
```

There's our data, let's now sort the data in descending order so we only see the top candidates

```python
jobdata = jobdata.sort_values(by = 'Jobs', ascending = False)
jobdata.head(10)
```

It looks like Central Business District is a clear outlier, so lets get rid of it. Nobody wants to live in a business district anyways

```python
jobdata = jobdata[jobdata['Jobs'] < 80000] #select only the rows that have less than 80,000
```

Now lets actually plot a bar chart to more easily interpret the data

```python
plt.style.use('seaborn-bright')
jobdata.head(25).plot(kind = 'barh', title = "Amount of Jobs per Neighborhood",rot=5,fontsize=7) 
```

We can now see the top 25 rankings in terms of Jobs in Pittsburgh here, it looks like both North and West Oakland are great for finding work. But let us also see the liklihood of also having to commute to work because no one really wants to travel to there job. For instance, in 2018 a study found that the average commute to work takes up to 27 minutes a person's time one way. [1] This works out to around 225 hours a year dedicated to getting to work. If we can work from home, its attractiveness for housing is much higher


```python
commutedata = data[['SNAP_All_csv_Commute_to_Work__D']] 
commutedata.dropna()
#drop all missing values

for row in commutedata:
    try:
        commutedata['SNAP_All_csv_Commute_to_Work__D'] = commutedata['SNAP_All_csv_Commute_to_Work__D'].str.rstrip('%').astype('float')
        #strip away the % from string and convert to float 
    except:
        pass
commutedata = commutedata.sort_values(by = 'SNAP_All_csv_Commute_to_Work__D', ascending = True) #get the smallest percentages

plt.style.use('ggplot')
commutedata.columns = {'Commute Rate in %'} 
commutedata.head(10).plot(kind = 'barh', rot = 5,fontsize=6, title = 'Liklihood of Commute per Neighborhood')
```
