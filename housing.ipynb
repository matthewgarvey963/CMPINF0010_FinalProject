```python
import pandas as pd
import numpy as np

housing_pandas = pd.read_csv("housing.csv")
housing_pandas.head()
```

```python
housing_pandas = housing_pandas[['Neighborhood', 'Median Home Value (2010)']]
#housing_pandas["Median Home Value (2010)"] = housing_pandas["Median Home Value (2000)"].str.replace(',', '')
#housing_pandas["Median Home Value (2010)"] = housing_pandas["Median Home Value (2000)"].str.replace('$', '')
#housing_pandas["Median Home Value (2010)"].mean()
```

```python
housing_pandas['Median Home Value (2010)'] = housing_pandas['Median Home Value (2010)'].str.replace(',', '')
housing_pandas['Median Home Value (2010)'] = housing_pandas['Median Home Value (2010)'].str.replace('$', '')
housing_pandas['Median Home Value (2010)'] = pd.to_numeric(housing_pandas['Median Home Value (2010)'], errors='coerce')
housing_pandas.head(100)
```

```python
housing_pandas.plot(x='Neighborhood', kind = 'bar', rot = 90, figsize = (30,10))
```

While rather large, this graph represents the median home value in all neighborhoods in Pittsburgh.

```python
middleClass = housing_pandas.loc[100000 < housing_pandas['Median Home Value (2010)']]
middleClass.plot(x='Neighborhood', kind = 'bar', rot = 90, figsize = (30,20))
```

Now we're able to see affordable housing for the middle-class and the graph shows areas of affortability for most individuals that fall into this category.


```python
lowerClass1 = housing_pandas.loc[100000 > housing_pandas['Median Home Value (2010)']]
lowerClass1.plot(x='Neighborhood', kind = 'bar', rot = 90, figsize = (30,20))
```

This chart represents meadian housing under 100,000. This is targeting the lower class and provides a wide variety of neighborhoods individuals can choose from.

```python
UpperClass = housing_pandas.loc[200000 < housing_pandas['Median Home Value (2010)']]
UpperClass.plot(x='Neighborhood', kind = 'bar', rot = 90, figsize = (30,20))
```

Based on this graph, clearly this is targeting the upper class which there only a select number of neighborhoods to chose from. If the best neighborhood to move to was based on median housing price, Squirrel Hill North would be the best location.
